```python
import numpy as np
print("Original: ", np.array([True, 1, 2]) + np.array([3, 4, False]))
# first array is 1,1,2 because true equivalent to 1, all elements coerced to integer type
# second array is 3,4,0 because false equivalent to 0
# now adding them by corresponding index numbers to get [4,5,2]
print("Answer: ", np.array([4, 3, 0]) + np.array([0, 2, 2]))
# essentially doing same thing, just add corresponding indexes and you will get same values
```
