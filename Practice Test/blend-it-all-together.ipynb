```python
import numpy as np
positions = ['GK', 'M', 'A', 'D', 'A', 'A', 'D', 'D', 'GK', 'GK', 'GK']
heights = [191,184,185,180,176,181,182,183,187,186,184]
np_positions = np.array(positions)
np_heights = np.array(heights)
print("Positions of players: ", np_positions)
print("Heights of players: ", np_heights)
gk_heights = np_heights[np_positions == 'GK']
other_heights=np_heights[np_positions!='GK']
print("Heights of goalkeepers: ", gk_heights)
print("Heights of other players: ", other_heights)
gk_median = np.median(gk_heights)
other_median = np.median(other_heights)
print("Median of heights of goalkeepers: ", gk_median)
print("Median of heights of other players: ", other_median)
print("Proven that median height of goalkeepers is higher!")
```
